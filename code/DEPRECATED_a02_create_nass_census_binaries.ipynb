```python
gz_path = Path("../Data/census_of_agriculture/qs.census2002.txt.gz")
census_df = pd.read_csv(gz_path, compression='gzip', sep='\t', skiprows=0, header=0, dtype=census_2002_dtype_dict, parse_dates=['LOAD_TIME'])
```
